```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("THE World University Rankings 2016-2024.csv")
df
```

```python
df_sinDuplicados= df.drop_duplicates()

```

```python
df= df_sinDuplicados
df.isnull()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Overall Score'], bins=30, kde=False, color='skyblue')
plt.title('Distribución de Clasificaciones de Universidades')
plt.xlabel('Clasificación Mundial')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
df['Country'].value_counts().plot(kind='bar', color='salmon')
plt.title('Proporción de Universidades por País')
plt.xlabel('País')
plt.ylabel('Número de Universidades')
plt.show()
```

```python
df=df.sort_values('Overall Score', ascending=False)
plt.figure(figsize=(12, 8))
sns.histplot(x='Country', y='Overall Score', data=df.head(100), palette='Set2')
plt.title('Distribución de Clasificaciones por País')
plt.xlabel('País')
plt.ylabel('Clasificación Mundial')
plt.show()
```

```python
df=df.sort_values('Overall Score', ascending=False)
plt.figure(figsize=(12, 8))
sns.histplot(x='Country', y='Industry Impact', data=df.head(100), palette='Set2')
plt.title('Distribución de Clasificaciones por impacto en la industria')
plt.xlabel('País')
plt.ylabel('Enseñanza')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Research Quality'], bins=30, kde=False, color='purple')
plt.title('Distribución de Puntuaciones de Investigación')
plt.xlabel('Puntuación de Investigación')
plt.ylabel('Frecuencia')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Research Quality', y='Overall Score', data=df, color='red')
plt.title('Relación entre Puntuación de Investigación y Clasificación Mundial')
plt.xlabel('Puntuación de Investigación')
plt.ylabel('Clasificación Mundial')
plt.show()
```

```python
df= df.sort_values(['Teaching'], ascending=True)
plt.figure(figsize=(12, 8))
sns.boxplot(x='Name', y='Teaching', data=df.head(30), palette='Blues')
plt.title('Puntuación de Enseñanza por País')
plt.xlabel('País')
plt.ylabel('Puntuación de Enseñanza')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Teaching', y='Research Quality', hue='Overall Score', data=df, palette='viridis', size='Overall Score')
plt.title('Puntuación de Enseñanza y Puntuación de Investigación por Clasificación Mundial')
plt.xlabel('Puntuación de Enseñanza')
plt.ylabel('Puntuación de Investigación')
plt.legend(title='Clasificación Mundial')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Research Environment', y='Research Quality', data=df, color='purple')
plt.title('Relación entre Puntuación de Investicacion medioambiental y Puntuación de Investigación')
plt.xlabel('Investicacion medioambiental')
plt.ylabel('Puntuación de Investigación')
plt.show()
```
