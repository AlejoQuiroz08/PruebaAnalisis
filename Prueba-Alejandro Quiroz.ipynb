```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("THE World University Rankings 2016-2024.csv")
df
```

```python
df_sinDuplicados= df.drop_duplicates()

```

```python
df= df_sinDuplicados
df.isnull()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Overall Score'], bins=30, kde=False, color='skyblue')
plt.title('DistribuciÃ³n de Clasificaciones de Universidades')
plt.xlabel('ClasificaciÃ³n Mundial')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
df['Country'].value_counts().plot(kind='bar', color='salmon')
plt.title('ProporciÃ³n de Universidades por PaÃ­s')
plt.xlabel('PaÃ­s')
plt.ylabel('NÃºmero de Universidades')
plt.show()
```

```python
df=df.sort_values('Overall Score', ascending=False)
plt.figure(figsize=(12, 8))
sns.histplot(x='Country', y='Overall Score', data=df.head(100), palette='Set2')
plt.title('DistribuciÃ³n de Clasificaciones por PaÃ­s')
plt.xlabel('PaÃ­s')
plt.ylabel('ClasificaciÃ³n Mundial')
plt.show()
```

```python
df=df.sort_values('Overall Score', ascending=False)
plt.figure(figsize=(12, 8))
sns.histplot(x='Country', y='Industry Impact', data=df.head(100), palette='Set2')
plt.title('DistribuciÃ³n de Clasificaciones por impacto en la industria')
plt.xlabel('PaÃ­s')
plt.ylabel('EnseÃ±anza')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Research Quality'], bins=30, kde=False, color='purple')
plt.title('DistribuciÃ³n de Puntuaciones de InvestigaciÃ³n')
plt.xlabel('PuntuaciÃ³n de InvestigaciÃ³n')
plt.ylabel('Frecuencia')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Research Quality', y='Overall Score', data=df, color='red')
plt.title('RelaciÃ³n entre PuntuaciÃ³n de InvestigaciÃ³n y ClasificaciÃ³n Mundial')
plt.xlabel('PuntuaciÃ³n de InvestigaciÃ³n')
plt.ylabel('ClasificaciÃ³n Mundial')
plt.show()
```

```python
df= df.sort_values(['Teaching'], ascending=True)
plt.figure(figsize=(12, 8))
sns.boxplot(x='Name', y='Teaching', data=df.head(30), palette='Blues')
plt.title('PuntuaciÃ³n de EnseÃ±anza por PaÃ­s')
plt.xlabel('PaÃ­s')
plt.ylabel('PuntuaciÃ³n de EnseÃ±anza')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Teaching', y='Research Quality', hue='Overall Score', data=df, palette='viridis', size='Overall Score')
plt.title('PuntuaciÃ³n de EnseÃ±anza y PuntuaciÃ³n de InvestigaciÃ³n por ClasificaciÃ³n Mundial')
plt.xlabel('PuntuaciÃ³n de EnseÃ±anza')
plt.ylabel('PuntuaciÃ³n de InvestigaciÃ³n')
plt.legend(title='ClasificaciÃ³n Mundial')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Research Environment', y='Research Quality', data=df, color='purple')
plt.title('RelaciÃ³n entre PuntuaciÃ³n de Investicacion medioambiental y PuntuaciÃ³n de InvestigaciÃ³n')
plt.xlabel('Investicacion medioambiental')
plt.ylabel('PuntuaciÃ³n de InvestigaciÃ³n')
plt.show()
```
